```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('../data/Online Retail.xlsx', nrows = 50000,dtype = {'StockCode':str})
df
```

#### bar chart top 10 quantity

```python
fig = plt.figure()
ax = fig.add_subplot()
quantity = df['Description'].value_counts().head(10)
ax.bar(quantity.index,quantity.values)

ax.set_xlabel('Quantity')
ax.set_ylabel('InvoiceDate')
ax.set_title('Top 10 Quantity')
plt.xticks(rotation = 90)
plt.show()
```

```python
quantity = df['Description'].value_counts().head(10).to_frame()
quantity
```

```python
quantity = df['UnitPrice'].groupby(df['StockCode']).sum()
quantity.sort_values(ascending=False).head(10)
```

```python
fig = plt.figure()
ax = fig.add_subplot()

quantity = df['UnitPrice'].groupby(df['StockCode']).sum()
top_10_quantity = quantity.sort_values(ascending=False).head(10)
ax.bar(top_10_quantity.index,top_10_quantity.values)

ax.set_xlabel('Stock')
ax.set_ylabel('Price')
ax.set_title('Top 10 Quantity')
plt.xticks(rotation = 90)
plt.show()
```
